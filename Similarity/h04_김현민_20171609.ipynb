```python
def jaccard_similarity(s1, s2):
    if len(s1|s2) == 0: # 합집합 연산
        return 0
    return len(s1&s2)/len(s1|s2) #교집합 갯수/ 합집합 갯수
```

```python
def pearson_similarity(u1, u2):
    u1_c = u1 - u1.mean() #평균으로 빼줌
    u2_c = u2 - u2.mean() #평균으로 빼줌
    denom = (np.sum(u1_c ** 2) * (u2_c ** 2).sum()) ** 0.5 #root 씌워주기
    if denom != 0:
        return (u1_c * u2_c).sum()/denom
    else:
        return 0
```

```python
def find_similar_movies (input_title, n, alpha):
    input_meta = meta.loc[ meta[ 'title'] == input_title].iloc[ 0] #여러개가 검색되면 처음꺼 가져옴
    input_set = input_meta.genres | input_meta.keywords
    
    result = []
    
    for this_title in matrix.columns:
        if this_title == input_title: #자기자신은 제외!
            continue
        
        this_meta = meta.loc[ meta[ 'title'] == this_title].iloc[ 0]
        this_set = this_meta.genres | this_meta.keywords
        
        jaccard = jaccard_similarity(this_set, input_set)
        pearson = pearson_similarity(matrix[this_title], matrix[input_title])
        
        
        score = alpha * pearson + ( 1-alpha) * jaccard
        result.append( (this_title, pearson, jaccard, score) )
        
    result.sort(key= lambda r: r[3], reverse= True) #score를 기준으로 내림차순. -r[3]으로 써도됌
    
    
    return result[:n]
```

```python
result = find_similar_movies('Godzilla', 10, 0.3)
result = pd.DataFrame(result, columns=['title', 'pearson', 'jaccard','score'])
result
```
